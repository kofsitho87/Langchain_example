```python
from langchain_community.utilities import SerpAPIWrapper
import os
from dotenv import load_dotenv
import requests
import pandas as pd

load_dotenv()
```

```python
class SearchTools():
  # @tool("Search the internet")
  def search_internet(query):
    """Useful to search the internet 
    about a a given topic and return relevant results"""
    top_result_to_return = 4
    url = "https://serpapi.com/search"
    params = {"engine": "google_shopping", "q": query, "gl": "kr", "hl": "ko", "api_key": os.environ["SERPAPI_API_KEY"]}
    headers = {
        # 'X-API-KEY': os.environ['SERPAPI_API_KEY'],
        'content-type': 'application/json'
    }
    response = requests.request("GET", url, headers=headers, params=params)
    results = response.json()
    # results = response.json()['organic']
    # string = []
    # for result in results[:top_result_to_return]:
    #   try:
    #     string.append('\n'.join([
    #         f"Title: {result['title']}", f"Link: {result['link']}",
    #         f"Snippet: {result['snippet']}", "\n-----------------"
    #     ]))
    #   except KeyError:
    #     next

    # return '\n'.join(string)
    return results
```

```python
# From now on, Load the data from local csv file
df = pd.read_csv("assets/shopping-results/맥북 프로 15인치.csv")
```

```python
# Change the price column to string to integer
df['price'] = df['price'].replace({'₩': '', ',': ''}, regex=True).astype(int)
df['price'][:5]
```

```python
from langchain.chat_models import ChatOpenAI

model = ChatOpenAI(model="gpt-4", temperature=0)
```

```python
# drop unnessary columns for reducing used gpt token
df.drop(columns = ["link","product_link","product_id","serpapi_product_api", "serpapi_product_api", "thumbnail"])
df.columns
```

```python

from langchain_experimental.agents import create_pandas_dataframe_agent

#create pandas agent
agent = create_pandas_dataframe_agent(model, df, verbose=True)
```

```python
agent.run("how many rows are there?")
```

```python
# Q: 가장 비싼 제품의 상품이름은?
agent.run("What is the name of the cheapest product?")
```

```python
# evaluate

index_of_highest = df['price'].idxmin()
df.loc[index_of_highest]['title']
```

```python
# Q: 가장 비싼 제품의 링크 정보를 알려주세요
agent.run("Please let me know the link information of the most expensive product")
```

```python
# evaluate

index_of_highest = df['price'].idxmax()
df.loc[index_of_highest]["link"]
```

```python
# Q: 중고상품 중에 가격이 가장 저렴한 상품의 가격은 얼마야?
agent.run("What is the price of the cheapest used product?")
```

```python
# evaluate

# 'second_hand_condition'이 "중고"인 행들만 필터링
used_products = df[df['second_hand_condition'] == '중고']

# 'price'가 가장 낮은 행 찾기
lowest_price_row = used_products[used_products['price'] == used_products['price'].min()]

lowest_price_row
```
